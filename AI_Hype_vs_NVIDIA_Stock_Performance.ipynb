# Does AI Hype Drive NVIDIA Stock Performance?

Automated data pipeline combining financial market data (yfinance) and search interest data (Google Trends API) for 2019 - 2026 analysis of AI-related market sentiment.

In the analysis below, we will check if and how Nvidia's valuation is impacted by artificial intelligence, semiconductors and Nvidia Google searches. NVIDIA was named as a bet on how big the AI market will grow. Since its products are highly correlated with AI, there was a need to add a benchmark. We have chosen Microsoft stock since it's less volatile and more immune to AI hype.
Apart from Nvidia and artificial intelligence searches, we have also added semiconductors.

Variables used:
*  NVDA Stock Daily Close Price - converted to monthly percentage change
*  MSFT Stock Daily Close Price - converted to monthly percentage change
*  Nvidia Google Trends - converted to monthly percentage change
*  Artificial intelligence Google Trends - converted to monthly percentage change
* Semiconductors Google Trends - converted to monthly percentage change

# **2. Data collection**

## **2.1. Yahoo Finance via yfinance**


```python
NVDA = yf.download('NVDA', start='2018-12-31', end='2026-06-26')
MSFT = yf.download('MSFT', start='2018-12-31', end='2026-06-26')
```

## **2.2. Google Trends via pytrends:**

```python
pytrends = tr.TrendReq(hl='en-US',tz=360)
keywords = ['NVIDIA', 'artificial intelligence', 'semiconductors']
pytrends.build_payload(keywords, timeframe='2018-12-01 2026-06-26', geo='')
trends = pytrends.interest_over_time()
trends.head()
```

# **4. Data preparation**

## **4.1. Stocks data resampling**

For further analysis, NVDA and MSFT data need to be converted from daily to monthly. Month-end market valuation will be represented by the close price recorded on the last trading day of each month. For the purpose of clarity and in alignment with trends, the dataset month start date will be used as an index.

```python
NVDA_monthly = NVDA.resample('MS').last()
MSFT_monthly = MSFT.resample('MS').last()
```

```python
NVDA_monthly.shape
```

```python
MSFT_monthly.shape
```

## **4.2. Monthly change calculation**

```python
def pct_change_calculation(df, columnname, new_columnname):
  df[new_columnname] = df[columnname].pct_change()
  print(df.head())
```

```python
pct_change_list = [
    [NVDA_monthly, 'Close', 'NVDA_change'],
    [MSFT_monthly, 'Close', 'MSFT_change'],
    [trends, 'NVIDIA', 't_NVIDIA'],
    [trends, 'artificial intelligence', 't_artificial intelligence'],
    [trends, 'semiconductors', 't_semiconductors']
]
```

```python
for n in pct_change_list:
  pct_change_calculation(n[0], n[1], n[2])
```

For both of the price stock datasets, we have calculated the percentage change based on the close price.

For the **trends** dataset, we needed to calculate the percentage change for all three columns. This calculation for the first row will return NaN, so it was dropped from the dataset. Due to very little interest in semiconductors at the beginning of the dataset, the percentage change was changed from NaN (result from dividing 0 by 0) to 0.

## **4.4. Combining datasets**

```python
trends_col = ['t_NVIDIA', 't_artificial intelligence', 't_semiconductors']
```

```python
df = pd.concat(
    [NVDA_monthly['NVDA_change'],
     MSFT_monthly['MSFT_change'],
     trends[trends_col]],
    axis=1
    )
```

```python
df_final = df.mul(100)
```

After successfully combining NVDA and MSFT stock price change with trends, the numbers were multiplied by 100 for a clearer view of the data.

```python
df_final
```

As expected, percentage change calculation for the first row returned NaN. It will be dropped from the dataset. Due to very little interest in semiconductors at the beginning of the chosen period, the percentage change will also be changed from NaN (result from dividing 0 by 0) to 0.

```python
df_final = df_final.iloc[1:]
df_final.fillna(0, inplace=True)
df_final.head()
```

## 5.5. Stocks Lag Analysis

In the prior part, we were analysing whether the change in trends has a direct response in the stock market. However, often there is a time lag between a rise in interest and actual movement from investors. Let's see if AI hype drives the stock price.

Since we will proceed with at least three similar operations let's define a few functions:

```python
def lag_calculation(df, columnname, lag):
  df[columnname] = df[columnname].shift(lag)
```

We are also checking trends vs. stock price, so there is no need for a whole correlation matrix.

```python
trends_list = ['t_NVIDIA', 't_artificial intelligence', 't_semiconductors']
```

```python
stocks_list = ['NVDA_change', 'MSFT_change']
```

For each correalation matrix we will also need a heatmap.

```python
def lag_corr_calculation(df, corr_df, heatmap_name):
  df.dropna(inplace=True)
  corr_df = df.corr()
  corr_df = corr_df.loc[trends_list, stocks_list]
  print(corr_df)
  heatmap_name = px.imshow(corr_df,
                      text_auto=".2f",
                      title='<b>Correlation Heatmap</b>',
                      color_continuous_scale='Rainbow',
                      zmin=-1,
                      zmax=1)
  heatmap_name.update_layout(title_x=0.5)

  heatmap_name.show()
```

### 5.5.1. One-Month Lag

Let's create necessary dataframes.

```python
df_final_1M = df_final.copy()
```

```python
corr_1M = pd.DataFrame()
heatmap_1M = px.imshow
```

```python
for stock in stocks_list:
  lag_calculation(df_final_1M, stock, -1)
```

```python
df_final_1M
```

```python
lag_corr_calculation(df_final_1M, corr_1M, heatmap_1M)
```

One-month lag analysis shows slight changes, but nothing notable.

### 5.5.2. Two-Month Lag

```python
df_final_2M = df_final.copy()
corr_2M = pd.DataFrame()
heatmap_2M = px.imshow
```

```python
for stock in stocks_list:
  lag_calculation(df_final_2M, stock, -2)
```

```python
df_final_2M
```

```python
lag_corr_calculation(df_final_2M, corr_2M, heatmap_2M)
```

Two-month lag analysis also shows slight changes signalling reverse correlation, most notably -0.3 correlation between NVIDIA searches and MSFT stock price.

### 5.5.3. Three-Month Lag

```python
df_final_3M = df_final.copy()
corr_3M = pd.DataFrame()
heatmap_3M = px.imshow
```

```python
for stock in stocks_list:
  lag_calculation(df_final_3M, stock, -3)
```

```python
df_final_3M
```

```python
lag_corr_calculation(df_final_3M, corr_3M, heatmap_3M)
```

For the three-month lag analysis, we are back to nothing notable.
